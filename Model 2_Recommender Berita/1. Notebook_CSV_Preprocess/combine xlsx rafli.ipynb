```python
import glob
import pandas as pd
import numpy as np
```

```python
dir = '../content/content_kemendikbud/'

excelfile = glob.glob(dir + '*.xlsx')

dfs = []

for file in excelfile:
    df = pd.read_excel(file)
    dfs.append(df)
combined_df = pd.concat(dfs, ignore_index=True)
combined_df.to_excel('article content kemendikbud.xlsx', index=False)
```
