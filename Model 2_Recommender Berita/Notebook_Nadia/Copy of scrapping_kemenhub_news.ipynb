```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import glob
```

```python
base_URL = "https://www.cnbcindonesia.com/search?query=cpns+kemenhub&p="

article_tags = []

for i in range(1, 9):
    URL = base_URL + "{}".format(i)
    page = requests.get(URL)
    #print(page.text)
    soup = BeautifulSoup(page.content, "html.parser")
    articles = soup.find_all("article")
    articles = articles[-10:]

    for article in articles:
        article_tags.append(article)
        #print(article, end="\n"*2)
    # print(article_tags)

url_list = []
title_list = []
img_list = []


for article_tag in article_tags:
    try:
        a_tag = article_tag.find('a')
        h2_tag = article_tag.find('h2')
        img_tag = article_tag.find('img')

        if a_tag is not None:
            url = a_tag['href']
            url_list.append(url)

        if h2_tag is not None:
            title = h2_tag.get_text()
            title_list.append(title)

        if img_tag is not None:
            img_url = img_tag['src']
            img_list.append(img_url)

    except Exception as e:
        print(f"Error processing article: {e}")

df = pd.DataFrame({
    'Judul Berita' : title_list,
    'Link Berita': url_list,
    'Link Gambar': img_list,

})
print(df)
```

# Kompas Kemenhub

```python
base_URL = "https://www.kompas.com/tag/cpns+kemenhub?page"
title_list = []
link_list = []
img_list = []

for i in range(1, 9):
    URL = f"{base_URL}={i}"
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    article_list = soup.find_all("div", {"class": "article__grid"})

    for article_tag in article_list:
        try:
            a_tag = article_tag.find('a')
            img_tag = article_tag.find('img')

            if a_tag is not None:
                url = a_tag['href']
                link_list.append(url)


            if img_tag is not None:
                img_url = img_tag['src']
                img_list.append(img_url)



            title = article_tag.find('a', {'class': 'article__link'}).text.strip()
            title_list.append(title)

        except Exception as e:
            print(f"Error processing article: {e}")

# Create DataFrame
df1 = pd.DataFrame({
    'Judul Berita': title_list,
    'Link Berita': link_list,
    'Link Gambar': img_list,
})

print(df1)
```

```python
count = 0
isi_berita_list = []
title_list = []
for i in range(len(link_kemenhub_kompas)):
    if count < 4:
        URL = link_kemenhub_kompas[i]
        #print(URL)
        page = requests.get(URL)
        soup = BeautifulSoup(page.content, "html.parser")

        h1_tag = soup.find('h1', class_='read__title')
        isi_artikel = soup.find('div', class_='read__content')

        if isi_artikel and h1_tag is not None:
            title = h1_tag.get_text().strip()
            title_list.append(title)

            p_content = ""
            paragraphs = isi_artikel.find_all('p')
            for paragraph in paragraphs:
                p_content += paragraph.get_text().strip() + " "
            isi_berita_list.append(p_content.strip())

            count += 1
        else:
            print("Article not found for URL:", URL)
    else:
        break

kompas_kemenhub_content = pd.DataFrame({
    'Judul Berita': title_list,
    'Content': isi_berita_list,
})

print(count)
```

# CNN Kemenhub

```python
base_URL = "https://www.cnnindonesia.com/search/?query=cpns+kemenkes&page"

article_tags = []

for i in range(1, 3):
    URL = f"{base_URL}={i}"
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    articles = soup.find_all("article")

    for article in articles:
        article_tags.append(article)

url_list = []
title_list = []
img_list = []
theme_list = []

for article_tag in article_tags:
    try:
        a_tag = article_tag.find('a')
        h2_tag = article_tag.find('h2')
        img_tag = article_tag.find('img')

        if a_tag is not None:
            url = a_tag['href']
            url_list.append(url)

        if h2_tag is not None:
            title = h2_tag.get_text()
            title_list.append(title)
            theme_list.append('Kemenkes')

        if img_tag is not None:
            img_url = img_tag['src']
            img_list.append(img_url)

    except Exception as e:
        print(f"Error processing article: {e}")

df = pd.DataFrame({
    'Judul Berita' : title_list,
    'Link Berita': url_list,
    'Link Gambar': img_list,
    'Tema Berita': theme_list
})

print(df)
```

# Detik Kemenhub

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_URL = "https://www.detik.com/search/searchall?query=cpns%20kemenhub&page"

article_tags = []

# Fetch pages
for i in range(1, 5):
    URL = f"{base_URL}={i}"
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    articles = soup.find_all("article", class_="list-content__item")

    for article in articles:
        article_tags.append(article)

url_list = []
title_list = []
img_list = []

for article_tag in article_tags:
    try:
        a_tag = article_tag.find('a')
        h3_tag = article_tag.find('h3', {'class':'media__title'})
        img_div = article_tag.find('div',{'class':'media__image'})

        if a_tag is not None:
            url = a_tag['href']
            url_list.append(url)

        if h3_tag is not None:
            title = h3_tag.get_text(strip=True)
            title_list.append(title)

        if img_div is not None:
            span_tag = img_div.find('span')
            if span_tag and 'style' in span_tag.attrs:
                style_attr = span_tag['style']
                img_url_start = style_attr.find('url(') + 4
                img_url_end = style_attr.find(')', img_url_start)
                img_url = style_attr[img_url_start:img_url_end].strip('";')
                img_list.append(img_url)

    except Exception as e:
        print(f"Error processing article: {e}")

# Ensure lists have the same length by filling with empty strings if needed
max_len = max(len(title_list), len(url_list), len(img_list))
title_list += [""] * (max_len - len(title_list))
url_list += [""] * (max_len - len(url_list))
img_list += [""] * (max_len - len(img_list))

df = pd.DataFrame({
    'Judul Berita' : title_list,
    'Link Berita': url_list,
    'Link Gambar': img_list,
})
```

```python
dt_detik_kemenhub = pd.read_csv("/content/drive/MyDrive/Capstone/detik_kemenhub.csv")
link_kemenhub_detik = list(dt_detik_kemenhub['Link Berita'])
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

count = 0
isi_berita_list = []
title_list = []
for i in range(len(link_kemenhub_detik)):
    if count < 40:
        URL = link_kemenhub_detik[i]
        try:
            page = requests.get(URL)
            if page.status_code == 200:
                soup = BeautifulSoup(page.content, "html.parser")

                h1_tag = soup.find('h1')
                isi_artikel = soup.find('div', {'class': 'detail__body-text itp_bodycontent'})

                if isi_artikel and h1_tag is not None:
                    children = isi_artikel.findChildren("p", recursive=False)
                    p_content = ""
                    title = h1_tag.get_text()
                    title_list.append(title.strip())
                    for child in children:
                        div_child = child.get_text()
                        p_content += div_child + " " + "{}".format(";")
                    isi_berita_list.append(p_content.strip())
                    count += 1
                else:
                    print(f"Article not found in URL: {URL}")
            else:
                print(f"Failed to retrieve URL: {URL} with status code: {page.status_code}")

        except Exception as e:
            print(f"An error occurred while processing URL: {URL} - {str(e)}")

        # Add delay to prevent rate limiting
        time.sleep(1)
    else:
        break

detik_kemenhub_content = pd.DataFrame({
    'Judul Berita': title_list,
    'Content': isi_berita_list,
})

print(count)
```
