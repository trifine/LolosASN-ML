## **Kemenkumham Kompas**

```python
base_URL = "https://www.kompas.com/tag/cpns+kemenkumham?page"
title_list = []
link_list = []
img_list = []

for i in range(1, 9):
    URL = f"{base_URL}={i}"
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    article_list = soup.find_all("div", {"class": "article__grid"})

    for article_tag in article_list:  # Iterate over article_list instead of article_tags
        try:
            a_tag = article_tag.find('a')
            img_tag = article_tag.find('img')

            if a_tag is not None:
                url = a_tag['href']
                link_list.append(url)  # Append URL to link_list


            if img_tag is not None:
                img_url = img_tag['src']
                img_list.append(img_url)


            # Extracting title from article__link class
            title = article_tag.find('a', {'class': 'article__link'}).text.strip()
            title_list.append(title)

        except Exception as e:
            print(f"Error processing article: {e}")

# Create DataFrame
df1 = pd.DataFrame({
    'Judul Berita': title_list,
    'Link Berita': link_list,
    'Link Gambar': img_list,
})

print(df1)
```

```python
df1.to_csv('cpns_kemenkumham_kompas.csv',index=False)
```

## **Kemenkumham CNBC**

```python
base_URL = "https://www.cnbcindonesia.com/search?query=cpns+kemenkumham&p="

article_tags = []

for i in range(1, 3):
    URL = base_URL + "{}".format(i)
    page = requests.get(URL)
    #print(page.text)
    soup = BeautifulSoup(page.content, "html.parser")
    articles = soup.find_all("article")
    articles = articles[-10:]

    for article in articles:
        article_tags.append(article)
        #print(article, end="\n"*2)
    # print(article_tags)

url_list = []
title_list = []
img_list = []


for article_tag in article_tags:
    try:
        a_tag = article_tag.find('a')
        h2_tag = article_tag.find('h2')
        img_tag = article_tag.find('img')

        if a_tag is not None:
            url = a_tag['href']
            url_list.append(url)

        if h2_tag is not None:
            title = h2_tag.get_text()
            title_list.append(title)

        if img_tag is not None:
            img_url = img_tag['src']
            img_list.append(img_url)

    except Exception as e:
        print(f"Error processing article: {e}")

df = pd.DataFrame({
    'Judul Berita' : title_list,
    'Link Berita': url_list,
    'Link Gambar': img_list,

})

df.to_csv('cpns_kemenkumham_cnbc.csv',index=False)
```

```python
dt_kemenkumham = pd.read_csv("cpns_kemenkumham_cnbc.csv")
link_berita_kemenkumham = list(dt_kemenkumham['Link Berita'])
```

```python
# 49
count = 0
isi_berita_list = []
title_list = []
n_links = len(link_berita_kemenkumham)

for i in range(n_links):
    if count < n_links:
        URL = link_berita_kemenkumham[i]
        #print(URL)
        page = requests.get(URL)
        soup = BeautifulSoup(page.content, "html.parser")

        h1_tag = soup.find('h1')
        isi_artikel = soup.find('div', {'class': 'detail_text'})
        if isi_artikel and h1_tag is not None:
            children = isi_artikel.findChildren("p" , recursive=False)
            p_content = ""
            title = h1_tag.get_text()
            title_list.append(title.strip())
            for child in children:
                div_child = child.get_text()
                p_content +=  div_child + " " + "{}".format(";")
            isi_berita_list.append(p_content.strip())
            count += 1
        else:
            print("article not found")
    else:
        break
kemenkumham_content = pd.DataFrame({
    'Judul Berita': title_list,
    'Content': isi_berita_list,
})
kemenkumham_content.to_excel('link_berita_kemenkumham_cnbc.xlsx',index=False)
print(count)
```

## **Kemenkumham Detik**

```python
base_URL = "https://www.detik.com/search/searchall?query=cpns%20kemenkumham&sortby=time&page={}"

article_tags = []

for i in range(1,21):
    URL = base_URL + "/{}".format(i)
    page = requests.get(URL)
    #print(page.text)
    soup = BeautifulSoup(page.content, "html.parser")
    articles = soup.find_all("article")
    articles = articles[-10:]
    if not articles:
        print(f"No articles found on page {i}")  # Debugging line
    for article in articles:
        article_tags.append(article)
        #print(article, end="\n"*2)
    # print(article_tags)

url_list = []
title_list = []
img_list = []

for article_tag in article_tags:
    a_tag = article_tag.find('a')
    h2_tag = article_tag.find('h2')
    img_tag = article_tag.find('img')
    # img_tag = img_tag['src']

    if a_tag is not None:
        url = a_tag['href']
        url_list.append(url)
        #print("URL:", url)

    if h2_tag is not None:
        title = h2_tag.get_text().strip()
        title_list.append(title)
        #print("Title:", title)

    if img_tag:
        img_url = img_tag.get('src')
        img_list.append(img_url)

df = pd.DataFrame({
    'Judul Berita' : title_list,
    'Link Berita': url_list,
    'Link Gambar': img_list
})

df.to_csv('cpns_kemenkumham_detik.csv',index=False)
```
