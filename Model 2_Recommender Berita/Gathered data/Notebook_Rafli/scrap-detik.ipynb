# KEMENDIKBUD, KEMENAG, KEMENSOS

- Import Libraries
- Get URL
- Extracts `h2` for title, `a` for links, and `img[src]` for image links
- Save it to excel


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import glob
```

## Page 1-20 KEMENDIKBUD

```python
base_URL = "https://www.detik.com/search/searchall?query=cpns+kemendikbud&siteid=2"

article_tags = []

for i in range(1,21):
    URL = base_URL + "/{}".format(i)
    page = requests.get(URL)
    #print(page.text)
    soup = BeautifulSoup(page.content, "html.parser")
    articles = soup.find_all("article")
    articles = articles[-10:]

    for article in articles:
        article_tags.append(article)
        #print(article, end="\n"*2)
    # print(article_tags)

url_list = []
title_list = []
img_list = []

for article_tag in article_tags:
    a_tag = article_tag.find('a')
    h2_tag = article_tag.find('h2')
    img_tag = article_tag.find('img')
    img_tag = img_tag['src']

    if a_tag is not None:
        url = a_tag['href']
        url_list.append(url)
        #print("URL:", url)

    if h2_tag is not None:
        title = h2_tag.get_text()
        title_list.append(title)
        #print("Title:", title)

    if img_tag is not None:
        img_list.append(img_tag)
    #print("URL:", url_list)
#print("Title:", title_list)

df = pd.DataFrame({
    'Judul Berita' : title_list,
    'Link Berita': url_list,
    'Link Gambar': img_list
})

df.to_csv('kemendikbud.csv',index=False)
```

# Contents Article KEMENDIKBUD

```python
dt_kemendikbud = pd.read_csv("kemendikbud.csv")
link_berita_kemendikbud = list(dt_kemendikbud['Link Berita'])
```

```python
# 49
count = 0
isi_berita_list = []
title_list = []
n_links = len(link_berita_kemendikbud)

for i in range(n_links):
    if count < n_links:
        URL = link_berita_kemendikbud[i]
        #print(URL)
        page = requests.get(URL)
        soup = BeautifulSoup(page.content, "html.parser")

        h1_tag = soup.find('h1', {'class': 'detail__title'})
        isi_artikel = soup.find('div', {'class': 'detail__body-text itp_bodycontent'})
        if isi_artikel and h1_tag is not None:
            children = isi_artikel.findChildren("p" , recursive=False)
            p_content = ""
            title = h1_tag.get_text()
            title_list.append(title.strip())
            for child in children:
                div_child = child.get_text()
                p_content +=  div_child + " " + "{}".format(";")
            isi_berita_list.append(p_content.strip())
            count += 1
        else:
            print("article not found")
    else:
        break
kemendikbud_content = pd.DataFrame({
    'Judul Berita': title_list,
    'Content': isi_berita_list,
})
kemendikbud_content.to_excel('link_berita_kemendikbud.xlsx',index=False)
print(count)
```
