- Import Libraries
- Get URL
- Extracts `h2` for title, `a` for links, and `img[src]` for image links
- Save it to excel


## Page test

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import glob
```

# Kemendikbud

```python
base_URL = "https://www.kompas.com/tag/cpns+kemendikbud?page"
title_list = []
link_list = []
img_list = []

for i in range(1, 21):
    URL = f"{base_URL}={i}"
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    article_list = soup.find_all("div", {"class": "article__grid"})

    for article_tag in article_list:  # Iterate over article_list instead of article_tags
        try:
            a_tag = article_tag.find('a')
            img_tag = article_tag.find('img')

            if a_tag is not None:
                url = a_tag['href']
                link_list.append(url)  # Append URL to link_list


            if img_tag is not None:
                img_url = img_tag['src']
                img_list.append(img_url)


            # Extracting title from article__link class
            title = article_tag.find('a', {'class': 'article__link'}).text.strip()
            title_list.append(title)

        except Exception as e:
            print(f"Error processing article: {e}")

# Create DataFrame
df1 = pd.DataFrame({
    'Judul Berita': title_list,
    'Link Berita': link_list,
    'Link Gambar': img_list,
})

df1.to_csv('kemendikbud.csv',index=False)
```

```python
dt_kemendikbud = pd.read_csv("kemendikbud.csv")
link_berita_kemendikbud = list(dt_kemendikbud['Link Berita'])
```

```python
count = 0
isi_berita_list = []
title_list = []
for i in range(len(link_berita_kemendikbud)):
    if count < len(link_berita_kemendikbud):
        URL = link_berita_kemendikbud[i]
        #print(URL)
        page = requests.get(URL)
        soup = BeautifulSoup(page.content, "html.parser")

        h1_tag = soup.find('h1', class_='read__title')
        isi_artikel = soup.find('div', class_='read__content')

        if isi_artikel and h1_tag is not None:
            title = h1_tag.get_text().strip()
            title_list.append(title)

            p_content = ""
            paragraphs = isi_artikel.find_all('p')
            for paragraph in paragraphs:
                p_content += paragraph.get_text().strip() + " "
            isi_berita_list.append(p_content.strip())

            count += 1
        else:
            print("Article not found for URL:", URL)
    else:
        break

kemendikbud_content = pd.DataFrame({
    'Judul Berita': title_list,
    'Content': isi_berita_list,
})

kemendikbud_content.to_excel('link_berita_kemendikbud.xlsx',index=False)
print(count)
```
