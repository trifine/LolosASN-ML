## Page 1-20 CPNS

```python
base_URL = "https://www.cnbcindonesia.com/search?query=cpns+kemenkes&p="

article_tags = []

for i in range(1,21):
    URL = base_URL + "{}".format(i)
    page = requests.get(URL)
    #print(page.text)
    soup = BeautifulSoup(page.content, "html.parser")
    articles = soup.find_all("article")
    articles = articles[-10:]

    for article in articles:
        article_tags.append(article)
        #print(article, end="\n"*2)
    # print(article_tags)

url_list = []
title_list = []
img_list = []
theme_list = []


for article_tag in article_tags:
    try:
        a_tag = article_tag.find('a')
        h2_tag = article_tag.find('h2')
        img_tag = article_tag.find('img')

        if a_tag is not None:
            url = a_tag['href']
            url_list.append(url)

        if h2_tag is not None:
            title = h2_tag.get_text()
            title_list.append(title)
            theme_list.append('Kemenkes')

        if img_tag is not None:
            img_url = img_tag['src']
            img_list.append(img_url)

    except Exception as e:
        print(f"Error processing article: {e}")

df = pd.DataFrame({
    'Judul Berita' : title_list,
    'Link Berita': url_list,
    'Link Gambar': img_list,
    'Tema Berita': theme_list

})
print(df)
df.to_csv('/content/drive/MyDrive/Capstone/cnbc_kemenkes.csv',index=False)
```

# Kompas Kemenkes

```python
base_URL = "https://www.kompas.com/tag/cpns+kemenkes?page"
tema_berita = []
title_list = []
link_list = []
img_list = []

for i in range(1, 9):
    URL = f"{base_URL}={i}"
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    article_list = soup.find_all("div", {"class": "article__grid"})

    for article_tag in article_list:  # Iterate over article_list instead of article_tags
        try:
            a_tag = article_tag.find('a')
            img_tag = article_tag.find('img')

            if a_tag is not None:
                url = a_tag['href']
                link_list.append(url)  # Append URL to link_list
                tema_berita.append('Kemenkes')


            if img_tag is not None:
                img_url = img_tag['src']
                img_list.append(img_url)


            # Extracting title from article__link class
            title = article_tag.find('a', {'class': 'article__link'}).text.strip()
            title_list.append(title)

        except Exception as e:
            print(f"Error processing article: {e}")

# Create DataFrame
df1 = pd.DataFrame({
    'Judul Berita': title_list,
    'Link Berita': link_list,
    'Link Gambar': img_list,
    'Tema Berita': tema_berita
})

print(df1)
```

```python
df1.to_csv('/content/drive/MyDrive/Capstone/kompas_kemenkes.csv',index=False)
```

# CNN Kemenkes

```python
base_URL = "https://www.cnnindonesia.com/search/?query=cpns+kemenkes&page"

article_tags = []

for i in range(1, 3):
    URL = f"{base_URL}={i}"
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    articles = soup.find_all("article")

    for article in articles:
        article_tags.append(article)

url_list = []
title_list = []
img_list = []
theme_list = []

for article_tag in article_tags:
    try:
        a_tag = article_tag.find('a')
        h2_tag = article_tag.find('h2')
        img_tag = article_tag.find('img')

        if a_tag is not None:
            url = a_tag['href']
            url_list.append(url)

        if h2_tag is not None:
            title = h2_tag.get_text()
            title_list.append(title)
            theme_list.append('Kemenkes')

        if img_tag is not None:
            img_url = img_tag['src']
            img_list.append(img_url)

    except Exception as e:
        print(f"Error processing article: {e}")

df = pd.DataFrame({
    'Judul Berita' : title_list,
    'Link Berita': url_list,
    'Link Gambar': img_list,
    'Tema Berita': theme_list
})

print(df)
```

# Detik kemenkes

```python
base_URL = "https://www.detik.com/search/searchall?query=cpns+kemenkes&siteid=2&sortby=time&result_type=latest&page"

article_tags = []

for i in range(1, 5):
    URL = f"{base_URL}={i}"
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    articles = soup.find_all("article")

    for article in articles:
        article_tags.append(article)

url_list = []
title_list = []
img_list = []

for article_tag in article_tags:
    try:
        a_tag = article_tag.find('a')
        h2_tag = article_tag.find('h2')
        img_tag = article_tag.find('img')

        if a_tag is not None:
            url = a_tag['href']
            url_list.append(url)

        if h2_tag is not None:
            title = h2_tag.get_text()
            title_list.append(title)

        if img_tag is not None:
            img_url = img_tag['src']
            img_list.append(img_url)

    except Exception as e:
        print(f"Error processing article: {e}")

df = pd.DataFrame({
    'Judul Berita' : title_list,
    'Link Berita': url_list,
    'Link Gambar': img_list,
})

```
