```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import glob
```

```python
base_URL = "https://www.cnbcindonesia.com/search?query=cpns+kemendagri&p="

article_tags = []

for i in range(1, 3):
    URL = base_URL + "{}".format(i)
    page = requests.get(URL)
    #print(page.text)
    soup = BeautifulSoup(page.content, "html.parser")
    articles = soup.find_all("article")
    articles = articles[-10:]

    for article in articles:
        article_tags.append(article)
        #print(article, end="\n"*2)
    # print(article_tags)

url_list = []
title_list = []
img_list = []


for article_tag in article_tags:
    try:
        a_tag = article_tag.find('a')
        h2_tag = article_tag.find('h2')
        img_tag = article_tag.find('img')

        if a_tag is not None:
            url = a_tag['href']
            url_list.append(url)

        if h2_tag is not None:
            title = h2_tag.get_text()
            title_list.append(title)

        if img_tag is not None:
            img_url = img_tag['src']
            img_list.append(img_url)

    except Exception as e:
        print(f"Error processing article: {e}")

df = pd.DataFrame({
    'Judul Berita' : title_list,
    'Link Berita': url_list,
    'Link Gambar': img_list,

})
print(df)
```

```python
df.to_csv('/content/drive/MyDrive/Capstone/cnbc_kemendagri.csv',index=False)
```

```python
dt_kemendagri_cnbc = pd.read_csv("/content/drive/MyDrive/Capstone/cnbc_kemendagri.csv")
link_kemendagri_cnbc = list(dt_kemendagri_cnbc['Link Berita'])
```

```python
count = 0
isi_berita_list = []
title_list = []
for i in range(len(link_kemendagri_cnbc)):
    if count < 15:
        URL = link_kemendagri_cnbc[i]
        #print(URL)
        page = requests.get(URL)
        soup = BeautifulSoup(page.content, "html.parser")

        h1_tag = soup.find('h1')
        isi_artikel = soup.find('div', {'class': 'detail_text'})
        if isi_artikel and h1_tag is not None:
            children = isi_artikel.findChildren("p" , recursive=False)
            p_content = ""
            title = h1_tag.get_text()
            title_list.append(title.strip())
            for child in children:
                div_child = child.get_text()
                p_content +=  div_child + " " + "{}".format(";")
            isi_berita_list.append(p_content.strip())
            count += 1
        else:
            print("article not found")
    else:
        break
cnbc_kemendagri_content = pd.DataFrame({
    'Judul Berita': title_list,
    'Content': isi_berita_list,
})

print(count)
```

# Kompas Kemendagri

```python
base_URL = "https://www.kompas.com/tag/cpns+kemendagri?page"
title_list = []
link_list = []
img_list = []

for i in range(1, 9):
    URL = f"{base_URL}={i}"
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    article_list = soup.find_all("div", {"class": "article__grid"})

    for article_tag in article_list:  # Iterate over article_list instead of article_tags
        try:
            a_tag = article_tag.find('a')
            img_tag = article_tag.find('img')

            if a_tag is not None:
                url = a_tag['href']
                link_list.append(url)  # Append URL to link_list


            if img_tag is not None:
                img_url = img_tag['src']
                img_list.append(img_url)


            # Extracting title from article__link class
            title = article_tag.find('a', {'class': 'article__link'}).text.strip()
            title_list.append(title)

        except Exception as e:
            print(f"Error processing article: {e}")

# Create DataFrame
df1 = pd.DataFrame({
    'Judul Berita': title_list,
    'Link Berita': link_list,
    'Link Gambar': img_list,
})

print(df1)
```

```python
df1.to_csv('/content/drive/MyDrive/Capstone/kompas_kemendagri.csv',index=False)
```

```python
dt_kompas_kemendagri = pd.read_csv("/content/drive/MyDrive/Capstone/kompas_kemendagri.csv")
link_kemendagri_kompas = list(dt_kompas_kemendagri['Link Berita'])
```

```python
count = 0
isi_berita_list = []
title_list = []
for i in range(len(link_kemendagri_kompas)):
    if count < 4:
        URL = link_kemendagri_kompas[i]
        #print(URL)
        page = requests.get(URL)
        soup = BeautifulSoup(page.content, "html.parser")

        h1_tag = soup.find('h1', class_='read__title')
        isi_artikel = soup.find('div', class_='read__content')

        if isi_artikel and h1_tag is not None:
            title = h1_tag.get_text().strip()
            title_list.append(title)

            p_content = ""
            paragraphs = isi_artikel.find_all('p')
            for paragraph in paragraphs:
                p_content += paragraph.get_text().strip() + " "
            isi_berita_list.append(p_content.strip())

            count += 1
        else:
            print("Article not found for URL:", URL)
    else:
        break

kompas_kemendagri_content = pd.DataFrame({
    'Judul Berita': title_list,
    'Content': isi_berita_list,
})

print(count)
```

# CNN Kemendagri

```python
base_URL = "https://www.cnnindonesia.com/search/?query=cpns+kemenkes&page"

article_tags = []

for i in range(1, 3):
    URL = f"{base_URL}={i}"
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    articles = soup.find_all("article")

    for article in articles:
        article_tags.append(article)

url_list = []
title_list = []
img_list = []
theme_list = []

for article_tag in article_tags:
    try:
        a_tag = article_tag.find('a')
        h2_tag = article_tag.find('h2')
        img_tag = article_tag.find('img')

        if a_tag is not None:
            url = a_tag['href']
            url_list.append(url)

        if h2_tag is not None:
            title = h2_tag.get_text()
            title_list.append(title)
            theme_list.append('Kemenkes')

        if img_tag is not None:
            img_url = img_tag['src']
            img_list.append(img_url)

    except Exception as e:
        print(f"Error processing article: {e}")

df = pd.DataFrame({
    'Judul Berita' : title_list,
    'Link Berita': url_list,
    'Link Gambar': img_list,
    'Tema Berita': theme_list
})

print(df)
```

# Detik Kemendagri

```python
base_URL = "https://www.detik.com/search/searchall?query=cpns+kemendagri&siteid=2&sortby=time&result_type=latest&page"

article_tags = []

for i in range(1, 5):
    URL = f"{base_URL}={i}"
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    articles = soup.find_all("article")

    for article in articles:
        article_tags.append(article)

url_list = []
title_list = []
img_list = []

for article_tag in article_tags:
    try:
        a_tag = article_tag.find('a')
        h2_tag = article_tag.find('h2')
        img_tag = article_tag.find('img')

        if a_tag is not None:
            url = a_tag['href']
            url_list.append(url)

        if h2_tag is not None:
            title = h2_tag.get_text()
            title_list.append(title)

        if img_tag is not None:
            img_url = img_tag['src']
            img_list.append(img_url)

    except Exception as e:
        print(f"Error processing article: {e}")

df = pd.DataFrame({
    'Judul Berita' : title_list,
    'Link Berita': url_list,
    'Link Gambar': img_list,
})
```
