# CPNS

- Import Libraries
- Get URL
- Extracts `h2` for title, `a` for links, and `img[src]` for image links
- Save it to excel


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import glob
```

## Page test

## **Kemenkominfo CNN**

```python
base_URL = "https://www.cnnindonesia.com/tag/cpns%20kemenkominfo"

article_tags = []

for i in range(1,21):
    URL = base_URL + "/{}".format(i)
    page = requests.get(URL)
    #print(page.text)
    soup = BeautifulSoup(page.content, "html.parser")
    articles = soup.find_all("article")
    articles = articles[-10:]

    for article in articles:
        article_tags.append(article)
        #print(article, end="\n"*2)
    # print(article_tags)

url_list = []
title_list = []
img_list = []

for article_tag in article_tags:
    a_tag = article_tag.find('a')
    h2_tag = article_tag.find('h2')
    img_tag = article_tag.find('img')
    img_tag = img_tag['src']

    if a_tag is not None:
        url = a_tag['href']
        url_list.append(url)
        #print("URL:", url)

    if h2_tag is not None:
        title = h2_tag.get_text()
        title_list.append(title)
        #print("Title:", title)

    if img_tag is not None:
        img_list.append(img_tag)
    #print("URL:", url_list)
#print("Title:", title_list)

df = pd.DataFrame({
    'Judul Berita' : title_list,
    'Link Berita': url_list,
    'Link Gambar': img_list
})

df.to_csv('cpns_kemenkominfo_cnn.csv',index=False)
```

```python
dt_kemenkominfo = pd.read_csv("cpns_kemenkominfo_cnn.csv")
link_berita_kemenkominfo = list(dt_kemenkominfo['Link Berita'])
```

```python
# 49
count = 0
isi_berita_list = []
title_list = []
n_links = len(link_berita_kemenkominfo)

for i in range(n_links):
    if count < n_links:
        URL = link_berita_kemenkominfo[i]
        #print(URL)
        page = requests.get(URL)
        soup = BeautifulSoup(page.content, "html.parser")

        h1_tag = soup.find('h1', {'class': 'mb-2 text-[28px] leading-9 text-cnn_black'})
        isi_artikel = soup.find('div', {'class': 'detail-text text-cnn_black text-sm grow min-w-0'})
        if isi_artikel and h1_tag is not None:
            children = isi_artikel.findChildren("p" , recursive=False)
            p_content = ""
            title = h1_tag.get_text()
            title_list.append(title.strip())
            for child in children:
                div_child = child.get_text()
                p_content +=  div_child + " " + "{}".format(";")
            isi_berita_list.append(p_content.strip())
            count += 1
        else:
            print("article not found")
    else:
        break
kemenkominfo_content = pd.DataFrame({
    'Judul Berita': title_list,
    'Content': isi_berita_list,
})
kemenkominfo_content.to_excel('link_berita_kemenkominfo_cnn.xlsx',index=False)
print(count)
```

## **Kemenkominfo kompas**

```python
base_URL = "https://www.kompas.com/tag/cpns+kemenkominfo?page"
title_list = []
link_list = []
img_list = []

for i in range(1, 9):
    URL = f"{base_URL}={i}"
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    article_list = soup.find_all("div", {"class": "article__grid"})

    for article_tag in article_list:  # Iterate over article_list instead of article_tags
        try:
            a_tag = article_tag.find('a')
            img_tag = article_tag.find('img')

            if a_tag is not None:
                url = a_tag['href']
                link_list.append(url)  # Append URL to link_list


            if img_tag is not None:
                img_url = img_tag['src']
                img_list.append(img_url)


            # Extracting title from article__link class
            title = article_tag.find('a', {'class': 'article__link'}).text.strip()
            title_list.append(title)

        except Exception as e:
            print(f"Error processing article: {e}")

# Create DataFrame
df1 = pd.DataFrame({
    'Judul Berita': title_list,
    'Link Berita': link_list,
    'Link Gambar': img_list,
})

print(df1)
```

```python
df1.to_csv('cpns_kemenkominfo_kompas.csv',index=False)
```

```python
# 49
count = 0
isi_berita_list = []
title_list = []

for i in range(len(link_berita_kemenkominfo)):
    if count < (len(link_berita_kemenkominfo)):
        URL = link_berita_kemenkominfo[i]
        #print(URL)
        page = requests.get(URL)
        soup = BeautifulSoup(page.content, "html.parser")

        h1_tag = soup.find('h1', class_='read__title')
        isi_artikel = soup.find('div', class_='read__content')
        if isi_artikel and h1_tag is not None:
            title = h1_tag.get_text().strip()
            title_list.append(title)

            p_content = ""
            paragraphs = isi_artikel.find_all('p')
            for paragraph in paragraphs:
                p_content += paragraph.get_text().strip() + " "
            isi_berita_list.append(p_content.strip())

            count += 1
        else:
            print("Article not found for URL:", URL)
    else:
        break
kemenkominfo_content = pd.DataFrame({
    'Judul Berita': title_list,
    'Content': isi_berita_list,
})
kemenkominfo_content.to_excel('link_berita_kemenkominfo_kompas.xlsx',index=False)
print(count)
```

## **Kemenkominfo Detik**

```python
base_URL = "https://www.detik.com/search/searchall?query=cpns%20kemenkominfo&sortby=time&page={}"

article_tags = []

for i in range(1,21):
    URL = base_URL + "/{}".format(i)
    page = requests.get(URL)
    #print(page.text)
    soup = BeautifulSoup(page.content, "html.parser")
    articles = soup.find_all("article")
    articles = articles[-10:]
    if not articles:
        print(f"No articles found on page {i}")  # Debugging line
    for article in articles:
        article_tags.append(article)
        #print(article, end="\n"*2)
    # print(article_tags)

url_list = []
title_list = []
img_list = []

for article_tag in article_tags:
    a_tag = article_tag.find('a')
    h2_tag = article_tag.find('h2')
    img_tag = article_tag.find('img')
    # img_tag = img_tag['src']

    if a_tag is not None:
        url = a_tag['href']
        url_list.append(url)
        #print("URL:", url)

    if h2_tag is not None:
        title = h2_tag.get_text().strip()
        title_list.append(title)
        #print("Title:", title)

    if img_tag:
        img_url = img_tag.get('src')
        img_list.append(img_url)

df = pd.DataFrame({
    'Judul Berita' : title_list,
    'Link Berita': url_list,
    'Link Gambar': img_list
})

df.to_csv('cpns_kemenkominfo_detik.csv',index=False)
```
